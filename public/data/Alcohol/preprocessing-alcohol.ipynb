```python
# Create a list to store data for the new CSV file
unique_locations = set()  # Set to store unique locations
data_2020_2021 = []

with open('alcohol.csv', newline='', encoding='utf-8-sig') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    rows = list(csv_reader)  # Store rows in a list

    # Access data using cleaned headers
    clean_headers = [header.lstrip('\ufeff') for header in csv_reader.fieldnames]

    # Loop through the stored rows for data processing
    for row in rows:
        location = row['LOCATION']  # Assuming 'LOCATION' is the cleaned header name

        # Check if the location is in the provided country codes list
        if location in country_codes:
            # Convert location short name to full name using the country_mapping dictionary
            full_name = country_mapping.get(location, location)

            # Update the row with the full country name
            row['LOCATION'] = full_name

            time = row['TIME']
            value = row['Value']

            # Perform operations using the extracted values
            print(f"Location: {full_name}, Time: {time},  Value: {value}")
            
            # Store or manipulate the data as needed
            data_2020_2021.append(row)

# Save the data to a new CSV file
with open('./filtered/filtered_alcohol.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=clean_headers)
    writer.writeheader()
    writer.writerows(data_2020_2021)

data_2020_2021 = pd.read_csv('./filtered/filtered_alcohol.csv')
# List of columns to drop
columns_to_drop = ['INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'Flag Codes']

# Drop the specified columns from the DataFrame
data_2020_2021 = data_2020_2021.drop(columns=columns_to_drop)

data_2020_2021.head()
data_2020_2021.to_csv('./filtered/filtered_alcohol.csv', index=False)
```

```python
import pandas as pd

# Define the function to reformat data
def reformat_data(input_file, output_file):
    # Read the CSV file
    data = pd.read_csv(input_file)

    # Perform the transformation
    formatted_data = data.melt(id_vars=['Countries, territories and areas', 'Beverage Types'],
                               var_name='year', value_name='value')

    # Rename columns
    formatted_data = formatted_data.rename(columns={
        'Countries, territories and areas': 'LOCATION',
        'Beverage Types': 'BEVERAGE TYPES',
        'year': 'TIME',
        'value': 'VALUE'
    })

    # Save the reformatted data to a new CSV file
    formatted_data.to_csv(output_file, index=False)

# Apply the function to each file
reformat_data('./filtered/filtered_alcohol_data_1960-1979.csv', './filtered/reformatted_alcohol_data_1960-1979.csv')
reformat_data('./filtered/filtered_alcohol_data_1980-1999.csv', './filtered/reformatted_alcohol_data_1980-1999.csv')
reformat_data('./filtered/filtered_alcohol_data_2000-2009.csv', './filtered/reformatted_alcohol_data_2000-2009.csv')
reformat_data('./filtered/filtered_alcohol_data_2011-2019.csv', './filtered/reformatted_alcohol_data_2011-2019.csv')
```
